```python
from time import time
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
```

```python
bubb_arr = []
insert_arr = []
selection_arr = []
merge_arr = []
quick_arr = []
```

> BUBBLE SORT
---

```python
for i in range(10,101):
    tic = time()
    for j in range(100):
        arr = np.random.randint(1,1000,i)
        Bubble(arr,len(arr))
    toc = time()
    bubb_arr.append((toc-tic)/100)
```

> INSERTION SORT
---

```python
for i in range(10,101):
    tic = time()
    for j in range(100):
        arr = np.random.randint(1,1000,i)  
        Insertion(arr,len(arr))
    toc = time()
    insert_arr.append((toc-tic)/100)
```

> SELECTION SORT
---


```python
for i in range(10,101):
     
    tic = time()
    for j in range(100):
        arr = np.random.randint(1,100,i) 
        SelectionSort(arr,len(arr))
    toc = time()
    selection_arr.append((toc-tic)/100)
```

> MERGE SORT
---

```python
for i in range(10,101):
     
    tic = time()
    for j in range(100):
        arr = np.random.randint(1,100,i) 
        IMergeSort(arr,len(arr))
    toc = time()
    merge_arr.append((toc-tic)/100)
```

> QUICK SORT
---

```python
for i in range(10,101):
    tic = time()
    for j in range(100):
        arr = np.random.randint(1,100000,100)
        QuickSort(arr,0,len(arr)-1)
    toc = time()
    quick_arr.append((toc-tic)/100)
```

```python
my_arr = np.array([bubb_arr,insert_arr,selection_arr,merge_arr,quick_arr]).transpose()
my_index = np.arange(10,101)
my_col = np.array(["Bubble","Insertion","Selection","Merge","Quick"])
```

```python
df = pd.DataFrame(my_arr,index=my_index,columns=my_col)
```
