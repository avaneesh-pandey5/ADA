Let there be an array of N random elements. We need to sort this array in ascending order. 
If n is very large (i.e. N= 1,00,000) then Quicksort may be considered as the fastest 
algorithm to sort this array. However, we can further optimize its performance by 
hybridizing it with insertion sort.
Therefore, if n is small (i.e. N<= 10) then we apply insertion sort to the array otherwise Quick Sort is applied. 
Implement the above discussed hybridized Quick Sort and compare the running time of 
normal Quick sort and hybridized quick sort. Run each type of sorting 10 times on a random 
set of inputs and compare the average time returned by these algorithms.
_____

#### Importing all the Modules

```python
from sorting_functions import *
import numpy as np
from time import time
import pandas as pd
import matplotlib.pyplot as plt
```

#### _Hybrid Sorting_
---

```python
hybrid = []
for i in range(100):
    toc = time()
    arr = np.arange(1,10*i)
    np.random.shuffle(arr)
    if len(arr) <= 100:
        Insertion(arr,len(arr))
    else:
        QuickSort(arr,0,len(arr)-1)
    tic = time()
    hybrid.append(tic-toc)
```

#### _Quick Sorting_
---

```python
normal = []
for i in range(100):
    toc = time()
    arr = np.arange(1,10*i)
    np.random.shuffle(arr)
    QuickSort(arr,0,len(arr)-1)
    tic = time()
    normal.append(tic-toc)
```

##### **Creating DataFrame from Pandas**

```python
my_arr = np.array([normal,hybrid]).transpose()
my_index = np.arange(1,101)
my_col = np.array(["Hybrid","Normal"])
```

```python
df = pd.DataFrame(my_arr,index=my_index,columns=my_col)
```

```python
df
```

##### **Plotting all the values of the DataFrame**

```python
df.plot(kind='line',figsize=(30,6))
```

##### **Plotting the sum of all the values**

```python
df_plot = df.sum(axis=0)
grapg_plot = df_plot.plot(kind='bar',figsize=(8,5),rot=0)

for i in grapg_plot.containers:
    grapg_plot.bar_label(i,label_type='edge',fontsize =10,padding=1)

plt.show()
```

After plotting the the graphs for a 100 runs we see that `Hybrid Quick Sort` performs slightly better than the `Normal Quick Sort`
